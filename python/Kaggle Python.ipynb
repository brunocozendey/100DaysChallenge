## Dicion√°rios

Os dicion√°rios s√£o estruturas de dados que mapeam valores em chaves.

```python
numbers = {'one':1, 'two':2, 'three':3}
```
Neste caso 'one', 'two', e 'three' s√£o chaves e 1, 2 e 3 s√£o seus valores correspondentes

Valores s√£o acessados via colchetes [], de forma similar as listas e strings

```python
numbers['one']
```

Podemos usar a mesma sintaxe para uma outra chave:

```python
numbers['eleven'] = 11
numbers
```
Ou podemos mudar o valor associado de uma chave existente:

```python
numbers['one'] = 'Pluto'
numbers
```

Python tem uma compreens√£o de sintax do dicion√°rio similar a compreens√£o de lista:

```python
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']
planet_to_initial = {planet: planet[0] for planet in planets}
planet_to_initial
```
O operador in nos diz se algo √© uma chave em um dicion√°rio.

```python
'Saturn' in planet_to_initial
'Betelgeuse' in planet_to_initial
```
Um loop for sobre um dicion√°rio ir√° retornar suas chaves:

```python
for k in numbers:
    print("{} = {}".format(k, numbers[k]))
```
Podemos acessar uma cole√ß√£o de todas as chaves ou de todos os valore usando os m√©todos: dict.keys() e dict.values(), respectivamente.

```python
# Get all the initials, sort them alphabetically, and put them in a space-separated string.
' '.join(sorted(planet_to_initial.values()))
```

Um m√©todo √∫til √© o dict.items(), que nos permite iterar sobre as chaves e valores de um dicion√°rio de form asimultanea. (No Python um item se refere ao par chave, valor)

```python
for planet, initial in planet_to_initial.items():
    print("{} begins with \"{}\"".format(planet.rjust(10), initial))
```
